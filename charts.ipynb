# Średnie kroczące i inne operacje w oknach

```python
import pandas as pd

df = pd.read_csv(
    './poker_hand.csv',
)
df['cumulative_win_loss'] = df['win_loss'].cumsum()
df['cumulative_win_loss'].plot(
    figsize=(15, 4),
    grid=True,
    title= 'Winnigs chart',
)
```
